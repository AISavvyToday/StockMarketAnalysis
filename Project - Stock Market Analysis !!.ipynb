```python
stocks = pd.DataFrame({'AAPL': apple['Adj Close'],\
                      'SBUX': sbux['Adj Close'],\
                      'WMT': wmt['Adj Close'],\
                      'GOOG': google['Adj Close'],\
                      'XOM': xom['Adj Close']})
```

```python
stocks.head()
```

```python
stocks.tail()
```

```python
stocks.plot(figsize=(10,6))
plt.show()
```

```python
stock_return = stocks.apply(lambda a: a/a[0])
```

```python
google['20d'] = google['Close'].rolling(window=20, center=False).mean()
```

```python
google['20d'].plot(figsize=(10,6))
```

```python
google['20d'].plot(figsize=(10,6))
google['Close'].plot(figsize=(10,6), color='red')
```

```python
apple['20d'] = apple['Close'].rolling(window=20, center=False).mean()
```

```python
apple['20d'].plot(figsize=(10,6), color='blue')
```

```python
apple['20d'].plot(figsize=(10,6), color='red')
apple['Close'].plot(figsize=(10,6), color='blue')
```

```python
# 50 days mov av
```

```python
apple['50d'] = apple['Close'].rolling(window=50, center=False).mean()
```

```python
apple['50d'].plot(figsize=(10,6), color='red')
apple['Close'].plot(figsize=(10,6), color='blue')
```

```python
# 20, 50, 100, 200
```

```python
google['50d'] = google['Close'].rolling(window=50, center=False).mean()
```

```python
google['100d'] = google['Close'].rolling(window=100, center=False).mean()
```

```python
google['200d'] = google['Close'].rolling(window=200, center=False).mean()
```

```python
google['Close'].plot(figsize=(10,6), color='blue',legend=True)
google['20d'].plot(figsize=(10,6), color='red', legend=True)
google['50d'].plot(figsize=(10,6), color='green', legend=True)
google['100d'].plot(figsize=(10,6), color='orange', legend=True)
google['200d'].plot(figsize=(10,6), color='purple', legend=True)
```

```python
google['Close'].plot(figsize=(10,6), color='blue',legend=True)

google['200d'].plot(figsize=(10,6), color='purple', legend=True)
```

```python
xom['20d'] = xom['Close'].rolling(window=20, center=False).mean()
```
